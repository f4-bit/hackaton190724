## Articles Scraping

#### eltiempo scraping

```python
from Scraping.eltiempo_scraping import ElTiempoScraper

eltiempo = ElTiempoScraper()
articles_eltiempo = eltiempo.obtener_articulos(paginas=100)
articles_eltiempo
```

```python
articles_eltiempo['content'] = None
articles_eltiempo['#coments'] = None
articles_eltiempo['coments'] = None
articles_eltiempo['tags'] = None
for index, row in articles_eltiempo.iterrows():
    content, coments, tags = eltiempo.get_information_eltiempo(row['url'])
    # Agregar el contenido del articulo
    articles_eltiempo.loc[index, 'content'] = content
    if coments is not None:
        articles_eltiempo.loc[index, '#coments'] = len(coments)
        articles_eltiempo.loc[index, 'coments'] = coments
    if tags is not None:
        articles_eltiempo.loc[index, 'tags'] = tags
```

```python
articles_eltiempo.head(10)
```

```python
# se guardan en un excel
articles_eltiempo.to_excel('./data/articles_eltiempo.xlsx',index=False)
```
